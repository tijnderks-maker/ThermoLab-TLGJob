## Ontwerp

## Procedure
```{exercise}
### **Procedure voor het bepalen van de soortelijke warmte van water**

1.  **Voorbereiding**
    *   Vul een maatbeker met een bekende hoeveelheid water (bijvoorbeeld 500 g = 0,5 kg).
    *   Meet de massa van het water nauwkeurig met een weegschaal.
    *   Plaats de thermometer of temperatuursensor in het water en noteer de begintemperatuur.

2.  **Opstelling**
    *   Plaats het elektrisch verwarmingselement volledig in het water.
    *   Sluit het verwarmingselement aan op de voedingsbron.
    *   Zorg dat een roerder (indien beschikbaar) aanstaat om de warmte gelijkmatig te verdelen.

3.  **Verwarming**
    *   Zet de voedingsbron aan en stel het vermogen in (10 W).
    *   Start tegelijkertijd de stopwatch.
    *   Verwarm het water gedurende een vooraf bepaalde tijd (5 minuten).
    *   Noteer de temperatuur van het water elke 20 seconden.

4.  **Afronding**
    *   Zet de voedingsbron uit na de gekozen tijd.
    *   Noteer de eindtemperatuur van het water.
    *   Bereken de totale energie die is toegevoerd:  
        $$ Q = P \times t $$  
        waarbij ( P ) het elektrisch vermogen is (in watt) en ( t ) de tijd (in seconden).

5.  **Berekening van soortelijke warmte**
    *   Gebruik de formule:  
        $$ c = \frac{Q}{m \times \Delta T} $$  
        waarbij:
        *   ( Q ) = toegevoerde energie (in joule),
        *   ( m ) = massa van het water (in kg),
        *   ( \Delta T ) = temperatuurstijging (in °C).

6.  **Controle**
    *   Vergelijk de berekende waarde van ( c ) met de bekende waarde voor water (( c \approx 4180 ,\mathrm{J/(kg \cdot °C)} )).
```
